```python
import pandas as pd
import numpy as np
```

```python
file_path = "data.csv"
data = pd.read_csv(file_path, encoding="latin1")
```

```python
print(data.columns)
```

```python
x = data["no2"].dropna().values
```

```python
r = 102303563

a = 0.05 * (r % 7)
b = 0.3 * ((r % 5) + 1)

print("a_r =", a)
print("b_r =", b)
```

```python
z = x + a * np.sin(b * x)
```

```python

mu = np.mean(z)
sigma2 = np.var(z)
lam = 1 / (2 * sigma2)
c = np.sqrt(lam / np.pi)
```

```python
print("\nFINAL PARAMETERS TO SUBMIT:")
print("mu =", mu)
print("lambda =", lam)
print("c =", c)
```
